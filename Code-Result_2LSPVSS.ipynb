This experiments aims to validate the computational tractabililty of the extended strip packing model.

1. Data Set Generation
2. Required parts calculation (different materials classification)
3. Opt_matching (Using Two dimensional Level Strip Packing Problem) 
   *Remainder consideration X
4. Matching group result recording
5. 20-days Keep going.

```python
index = [['class%d'%(num) for num in range(1,11)]]
columns = ['LSP_single_AREA','LSP_single_TIME','LSP_single_GAP',
           'LSP_double1_AREA','LSP_double1_TIME','LSP_double1_GAP',
           'LSP_double2_AREA','LSP_double2_TIME','LSP_double2_GAP',
           'LSP_triple_AREA','LSP_triple_TIME','LSP_triple_GAP']
```

```python
df1 = pd.DataFrame(matrix, columns = columns, index = index)
```

```python
df1.loc['Total'] = df1.sum()
```

```python
df1
```

```python
for i in range(11):
    area = df1.iloc[i,0]
    time = df1.iloc[i,1]
    gap = df1.iloc[i,2]
    for j in range(4):
        df1.iloc[i,3*j] /= area
        df1.iloc[i,3*j+1] /= time
        df1.iloc[i,3*j+2] /= gap
```
