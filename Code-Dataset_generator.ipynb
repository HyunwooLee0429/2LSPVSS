## Various size Strip packing problem

This experiments aims to validate the computational tractabililty of the extended strip packing model.

1. Data Set Generation
2. Required parts calculation (different materials classification)
3. Opt_matching (Using Two dimensional Level Strip Packing Problem) 
   *Remainder consideration X
4. Matching group result recording
5. 20-days Keep going.

```python
import pickle
import numpy as np
```

```python
Datasets = {}
sizes = [60, 80, 100]

M_and_V = [1,2,3,4]
B_and_W = [5,6,7,8,9,10]

W = 100
H = 100
n = 40

W_set = [W, W, W, W, 10, 30, 40, 100, 100, 300]
items_width_lb = [W*2/3, 1, W*1/2, 1, 1, 1, 1, 1, 1, 1]
items_width_ub = [W, W*1/2, W, W*1/2, 10, 10, 35, 35, 100, 100]
items_height_lb = [1, H*2/3, H*1/2, 1, 1, 1, 1, 1, 1, 1]
items_height_ub = [H*1/2, H, H, H*1/2, 10, 10, 35, 35, 100, 100]
```

```python
Datasets['class1'] = {}
for size in sizes:
    Datasets['class1'][size] = {}
    for i in range(5):
        Datasets['class1'][size][i] = []
        for _ in range(size):
            if np.random.rand(1) <= 0.7:
                width = np.random.uniform(low = items_width_lb[0], high = items_width_ub[0])
                height = np.random.uniform(low = items_height_lb[0], high = items_height_ub[0])
                Datasets['class1'][size][i].append((width, height))
            elif np.random.rand(1) <= 0.8:
                width = np.random.uniform(low = items_width_lb[1], high = items_width_ub[1])
                height = np.random.uniform(low = items_height_lb[1], high = items_height_ub[1])
                Datasets['class1'][size][i].append((width, height))
            elif np.random.rand(1) <= 0.9:
                width = np.random.uniform(low = items_width_lb[2], high = items_width_ub[2])
                height = np.random.uniform(low = items_height_lb[2], high = items_height_ub[2])
                Datasets['class1'][size][i].append((width, height))
            elif np.random.rand(1) <= 1.0:
                width = np.random.uniform(low = items_width_lb[3], high = items_width_ub[3])
                height = np.random.uniform(low = items_height_lb[3], high = items_height_ub[3])
                Datasets['class1'][size][i].append((width, height))
```

```python
for num in range(5,11):
    Datasets['class%s'%str(num)] = {}    
    for size in sizes:
        Datasets['class%s'%str(num)][size] = {}    
        for i in range(5):
            Datasets['class%s'%str(num)][size][i] = []    
            for _ in range(size):
                width = np.random.uniform(low = items_width_lb[num-1], high = items_width_ub[num-1])
                height = np.random.uniform(low = items_height_lb[num-1], high = items_height_ub[num-1])
                Datasets['class%s'%str(num)][size][i].append((width, height))                                                             
```

```python
for num in range(1,11):
    for size in sizes:
        for i in range(5):
            Datasets['class%s'%str(num)][size][i] = sorted([data for data in Datasets['class%s'%str(num)][size][i]], key = lambda x:x[1], reverse=True)
```
